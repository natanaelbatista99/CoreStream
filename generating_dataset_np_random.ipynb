```python
import numpy as np
import pandas as pd

# Definindo as dimensões do conjunto de dados
#n_rows = [100000, 250000, 500000, 500000, 500000, 500000, 750000, 1000000]
#n_cols = [4, 4, 2, 4, 6, 8, 4, 4]

n_rows = [700001]
n_cols = [4]

for i in range(len(n_rows)):
    # Gerando um conjunto de dados aleatórios com números entre 0 e 1
    data = np.random.rand(n_rows[i], n_cols[i])
    
    # Arredonda os valores para o número de casas decimais desejado
    data = np.round(data, 5)

    # Convertendo o array numpy para um dataframe pandas
    df = pd.DataFrame(data)

    # Salvando o conjunto de dados em um arquivo CSV
    df.to_csv('datasets/' + str(n_rows[i] // 1000) + 'k_' + str(n_cols[i]) + 'd.csv', index=False, header=False)
```
