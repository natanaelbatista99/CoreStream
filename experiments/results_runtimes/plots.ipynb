#Código para pegar os csvs de todos os datasets do HASTREAM e gerar uma média e desvio padrao do tempo total de execução

```python
combined_data = []
x_has     = ['40k4d', '55k4d', '70k2d', '70k6d', '70k8d', '85k4d', '100k4d', 'covertype', 'kddcup', 'census']
y_has     = []
y_std_has = []
y_sum_has = []

for ss in x_has:

  for arquivo_csv in glob.glob('/content/drive/MyDrive/hpc/output/hdbstream_runtime_'+str(ss)+'/*.csv'):
    df = pd.read_csv(arquivo_csv)
    combined_data.append(df)

  #mean = combined_data['total'].mean()
  #std = combined_data['total'].std()
  sum_das_colunas   = pd.concat(combined_data).groupby(level=0)['total'].sum()
  media_das_colunas = pd.concat(combined_data).groupby(level=0)['total'].mean()
  std_das_colunas   = pd.concat(combined_data).groupby(level=0)['total'].std()

  y_std_has.append(std_das_colunas[0])
  y_has.append(media_das_colunas[0])
  y_sum_has.append(sum_das_colunas[0])

y_has     = np.array(y_has) / 60
y_std_has = np.array(y_std_has) / 60
y_sum_has = np.array(y_sum_has) / 60
```

#Código para pegar os dvs de runtime do CORESTREAM e gerar média e desvio padrao do tempo total de execução.

```python
combined_data = []
x_cor     = ['40k4d', '55k4d', '70k2d', '70k6d', '70k8d', '85k4d', '100k4d', 'covertype', 'kddcup', 'census']
y_cor     = []
y_std_cor = []
y_sum_cor = []

for ss in x_cor:

  for arquivo_csv in glob.glob('/content/drive/MyDrive/hpc/output/corestream_runtime_'+str(ss)+'/*.csv'):
    df = pd.read_csv(arquivo_csv)
    combined_data.append(df)

  #mean = combined_data['total'].mean()
  #std = combined_data['total'].std()
  sum_das_colunas   = pd.concat(combined_data).groupby(level=0)['total'].sum()
  media_das_colunas = pd.concat(combined_data).groupby(level=0)['total'].mean()
  std_das_colunas   = pd.concat(combined_data).groupby(level=0)['total'].std()

  y_std_cor.append(std_das_colunas[0])
  y_cor.append(media_das_colunas[0])
  y_sum_cor.append(sum_das_colunas[0])

y_cor     = np.array(y_cor) / 60
y_std_cor = np.array(y_std_cor) / 60
y_sum_cor = np.array(y_sum_cor) / 60
```

#Código para plotar o grafíco de comparação de tempo de execução total entre o hastream e corestream da média e desvio padrao.

```python
fig, ax = plt.subplots(1,2, figsize=(20, 5))

largura_barra = 0.5

pos1 = np.arange(len(x_cor))
pos2 = [x + largura_barra for x in pos1]

ax[0].bar(pos1, y_has,largura_barra, yerr=y_std_has, capsize=5, align='center', alpha=0.7, ecolor='black', color="red")
ax[0].bar(pos2, y_cor,largura_barra, yerr=y_std_cor, capsize=5, align='center', alpha=0.7, ecolor='black', color="blue")
ax[0].set_xlabel('Datasets', fontsize = 15)
ax[0].set_ylabel('Time(minutes)', fontsize = 15)
ax[0].set_xticks([x + largura_barra / 2 for x in pos1], x_cor)  # Configurar rótulos do eixo x
ax[0].legend(('hastream','corestream'),shadow=True, fontsize = 13)
#ax[0].set_title('', fontsize = 15)
ax[0].text(4.7, -0.8, '(a) Mean and Standard Deviation timestamps runtime',fontsize=15, ha='center', va='center')

ax[1].bar(pos1, y_sum_has,largura_barra, capsize=5, align='center', alpha=0.7, ecolor='black', color="red")
ax[1].bar(pos2, y_sum_cor,largura_barra, capsize=5, align='center', alpha=0.7, ecolor='black', color="blue")
ax[1].set_xlabel('Datasets', fontsize = 15)
ax[1].set_ylabel('Time(minutes)', fontsize = 15)
ax[1].set_xticks([x + largura_barra / 2 for x in pos1], x_cor)  # Configurar rótulos do eixo x
ax[1].legend(('hastream','corestream'), shadow=True, fontsize = 13)
#ax[1].set_title('Total sum timestamps runtime', fontsize = 15)
ax[1].text(4.7,-28,'(b) Total runtime',fontsize=15, ha='center', va='center')

fig.savefig("/content/drive/MyDrive/hpc/output/plots_images/runtime_bar.eps", format='eps', bbox_inches='tight')
```
